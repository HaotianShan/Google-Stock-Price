```python
import pandas as pd
df = pd.read_csv("/kaggle/input/google-stock-price/GOOG.csv")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date']) 
df = df.sort_values('Date') 

last_10_years = df[df['Date'] >= pd.to_datetime('2013-12-05')]

plt.figure(figsize=(14, 7))
for i in ['Open', 'High', 'Low', 'Close', 'Adj Close']:
    plt.plot(last_10_years['Date'], last_10_years[i], label=i, linewidth=0.7)  
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices over the Last 10 Years')
plt.legend()
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(last_10_years['Date'], last_10_years['Volume'], label='Volume', linewidth=0.5) 
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Stock Volume over the Last 10 Years')
plt.legend()
plt.show()
```
